## 1.2 N차원 배열의 데이터 타입

### 1.2.1 dtype

import numpy as np

```python
# 데이터형식을 지정하기 위해서 어레이함수의 인자로 dtype을 사용함
arr = np.array([1,2,3], dtype=float)
print(arr, arr.dtype)
```

```python
arr1 = np.array([1.1,2.2,3.3], dtype=int)
print(arr1, arr1.dtype)
```

```python
arr2 = np.array([1,0,0], dtype=bool)
print(arr2, arr2.dtype)
```

```python
#astype
arr3 = np.array([0,1,2,3,4,5])
print(arr3, arr3.dtype)

arr3 = arr3.astype(np.float32)
print(arr3, arr3.dtype)
```

```python
#데이터 타입이 혼재하는 경우
arr4 = np.array([0,1,2,3,4.4,"5five"])
print(arr4, arr4.dtype)
```

```python
arr5 = np.array([1,2,"64"], dtype=int)
print(arr5, arr5.dtype)
```

## 1.3 정해진 형식의 N차원 배열 생성하기

### 1.3.1 np.zeros()

```python
#np.zeros()
arr = np.zeros([2, 2])
print(arr)
```

### 1.3.2 np.ones()

```python
#np.ones()
arr = np.ones([3,5])
print(arr)
```

### 1.3.3 np.full()

```python
#np.full()
arr = np.full((2,3),5)
print(arr)            
```

### 1.3.4 np.eye()

```python
#np.eye() 대각 원소는 모두 1 나머지는 모두 0, k값은 대각원소의 발생지점
arr = np.eye(3,4,k=0)
print(arr)             

arr = np.eye(3,4,k=1)
print(arr)             

arr = np.eye(3,4,k=-1)
print(arr)             

arr = np.eye(3)
print(arr)             
```

### 1.3.5 np.___.like()

```python
arr = np.array([[1,2,3],
               [4,5,6]])

#zeros.like()
arr_z = np.zeros_like(arr)
print(arr_z)

#ones.like()
arr_0 = np.ones_like(arr)
print(arr_0)

#full_like()
arr_f = np.full_like(arr,10)
print(arr_f)
```

### 1.4.3 logspace

```python
# logspace
arr = np.linspace(1,10,10)
print(arr, end="\n\n")

arr = np.logspace(1,10,10, base=2) #10개의 원소, 베이스 로그의 밑수.
print(arr)

arr = np.logspace(1,10,10)
print(arr)
```

## 1.5 난수로 n차원 배열 생서하기

```python
import matplotlib.pyplot as plt
```

### 1.5.1 random.normal()

```python
# np.random.normal()
arr = np.random.normal(0, 1, 10) #정규분포 평균, 표준편차, 추출할 표본개수
print(arr)

arr = np.random.normal(0, 1, (2,3))
print(arr)
```

```python
arr = np.random.normal(0,1,10000000)
plt.hist(arr, bins=100)
plt.show()
```

### 1.5.2 random.rand()

```python
# np.random.rand()
arr = np.random.rand(1000) #0에서 1의 숫자를 균당한 비율로 추출
plt.hist(arr, bins=100)
plt.show
```

### 1.5.3 random.randn()

```python
# np.random.randn. 가우시안 표준정규분포에서
arr = np.random.randn(1000)
plt.hist(arr, bins=100)
plt.show
```

### 1.5.4 random.randint()

```python
# np.random.randint()

arr = np.random.randint(low = 1, high =5, size = 10) #high미만 까지 10개
print(arr)

arr = np.random.randint(low = 1, high =5, size = (3,4))
print(arr)
```

## 1.6 시드(Seed)값을 통한 난수 생성 제어

```python
arr = np.random.rand(10)
print("난수발생 1\n",arr)

arr = np.random.rand(10)
print("난수발생 2\n",arr)
```

```python
# random.seed
np.random.seed(1)
arr = np.random.rand(10)
print("난수발생 1\n",arr)

np.random.seed(1)
arr = np.random.rand(10)
print("난수발생 2\n",arr)
```
